```python
n = 1000
k = 3.0
c = 4.0
qv = 2.
t_u = (np.random.rand(n)-.5)*20
t_u_e = 0 #(np.random.rand(n)-.5)*1
t_c = t_u*k+c +qv*t_u**2
t_c_e = (np.random.rand(n)-.5)*1
t_c = t_c + t_c_e
t_u = t_u + t_u_e

t_c_f = torch.tensor(t_c, dtype=torch.float32)
t_u_f = torch.tensor(t_u, dtype=torch.float32)


n_samples = t_u_f.shape[0]
n_val = int(n_samples*.2)

shuffled_ind = torch.randperm(n_samples)

train_ind = shuffled_ind[:-n_val]
val_ind = shuffled_ind[-n_val:]
val_ind

t_c = t_c_f[train_ind]
t_u = t_u_f[train_ind]

t_c_v = t_c_f[val_ind]
t_u_v = t_u_f[val_ind]
```

```python
t_u
```

```python
def model(t_u, w, b):
    return t_u*w + b

def loss_fn(t_p, t_c):
    sq_diff = (t_p - t_c)**2
    return sq_diff.mean()

def dloss_fn(t_p, t_c):
    dfn = 2*(t_p-t_c)/t_p.size(0)
    return dfn

def dmod_w(t_u, w, b):
    return t_u

def dmod_b(t_u, w, b):
    return 1

def grad_fn(t_u, t_c,t_p, w,b):
    dloss = dloss_fn(t_p, t_c)
    dloss_w = dloss*dmod_w(t_u, w, b)
    dloss_b = dloss*dmod_b(t_u, w,b)
    return torch.stack([dloss_w.sum(), dloss_b.sum()])

w = torch.ones(())
b = torch.zeros(())
```

```python
t_p = model(t_u, w, b)
t_p
```

```python
loss_fn(t_p, t_c)
```

```python
l_r = .001

for i in range(100000):
    t_p = model(t_u, w, b)
    grad = grad_fn(t_u, t_c,t_p, w, b)
    w = w - grad[0]*l_r
    b = b - grad[1]*l_r

(w, b)
```

```python
def training_loop(n_epoch, learning_rate, params, t_u, t_c):
    sc = t_c.std()/t_u.std()
    t_u_n = sc*t_u
    for epoch in range(1, n_epoch+1):
        w, b = params
        t_p = model(t_u_n,w,b)
        loss = loss_fn(t_p,t_c)
        grad = grad_fn(t_u_n, t_c,t_p, w, b)
        params = params - learning_rate*grad

        print(f'Epoch {epoch}. Loss: {loss}')
    params[0]*=sc
    return params
```
